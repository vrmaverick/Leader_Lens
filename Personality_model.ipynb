# Leadership Quality Prediction 

We will be using RandomForestClassifier for Classification of How LeaderShip Quality is inter related with values of other attributes

# Generating a baseline model

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('Leader',axis=1)
Y = df['Leader']
np.random.seed(67)
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.3) 
```

```python
model = RandomForestRegressor(n_estimators=9)
```

```python
model.get_params()
```

```python
model.fit(X_train,Y_train);
```

```python
Y_predict = model.predict(X_test)
Y_predict
```

# Evaluating The Regression Model

**Evaluating on basis of Rsquare or Coefficient of Detrmination**

```python
a = model.score(X_test,Y_test) # needs optimization and more larger data set
a
```

```python
from sklearn.metrics import r2_score
print(f"Perfect model R2 score :{r2_score(Y_test,Y_test)}")
Y_test_mean = np.full(len(Y_test),Y_test.mean())
print(f"Poor Performance R2 score:{r2_score(Y_test,Y_test_mean)}")
```

**Using `cross_val_score()` for better evaluation** (Higher the value better the evaluation)

```python
from sklearn.model_selection import cross_val_score
```

```python
np.random.seed(67)
print(f'More accurate description by cross validation {cross_val_score(model,X,Y,cv=3).mean()}, as normal score can be misleading{a}') # scoring parameter = defualt(R^2)
```

```python
cvs_mse =cross_val_score(model,X,Y,cv=5,scoring="neg_mean_squared_error",) 
print(f'cross validated -ve mean-square-error : {cvs_mse.mean()}')
```

```python
cvs_mae =cross_val_score(model,X,Y,cv=5,scoring="neg_mean_absolute_error",) 
print(f'cross validated -ve mean-absolute-error : {cvs_mae.mean()}')
```

# Tuning hyperparameters

* `max_depth`
* `max_feature`
* `min_sample_leaf`
* `min_samples_split`
* `n_estimators`

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
model.get_params()
```

```python
grid = {"n_estimators":[10,100,200,500,1000,1200],
        "max_depth":[None,5,10,20,30],
        "max_features":[1.0,"sqrt",None],
        "min_samples_split":[2,4,6],
        "min_samples_leaf":[1,2,4]
}
```

```python
# already we have X,Y,C_tain,Y_train
model = RandomForestRegressor(n_jobs=None)
rf_model = RandomizedSearchCV(estimator=model,param_distributions=grid,cv=5,n_iter=15,verbose=2)
rf_model.fit(X_train,Y_train)
```

```python
rf_model.best_params_
```

```python
rf_pred = rf_model.predict(X_test) 
rf_pred
```

# we have got the best set of hyperparameters lets see the predictions

**Experimentation**

```python
mean_absolute_error(Y_test,rf_pred)
```

```python
mean_squared_error(Y_test,Y_predict)
```

```python
rf_model.best_score_ #not better than basline manually tuned model
```

**Now Using gridsearchCV and only keeping the nearby values of best hyperparametes given to reduc search space as the gridSearchCV fits all the combinations possible**

```python
grid2 = {"n_estimators":[10,100],
        "max_depth":[10,20,30],
        "max_features":[1.0,"sqrt"],
        "min_samples_split":[2,4,6],
        "min_samples_leaf":[1,2]}
```

```python
from sklearn.model_selection import GridSearchCV
gs_model = GridSearchCV(estimator=model,param_grid=grid2,cv=5,verbose=2)
gs_model.fit(X_train,Y_train)
```

```python
gs_model.best_params_
```

```python
gs_pred = gs_model.predict(X_test)
gs_pred
```

```python
print(f"MAE : {mean_absolute_error(Y_test,gs_pred)},MSE : {mean_squared_error(Y_test,gs_pred)},Score = {gs_model.best_score_}")
```
