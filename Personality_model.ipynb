# Leadership Quality Prediction 

We will be using RandomForestClassifier for Classification of How LeaderShip Quality is inter related with values of other attributes

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('Leader',axis=1)
Y = df['Leader']
np.random.seed(73)
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.3) 
```

```python
model = RandomForestRegressor()
```

```python
model.fit(X_train,Y_train);
```

```python
Y_predict = model.predict(X_test)
Y_predict
```

```python
model.score(X_test,Y_test) # needs optimization and more larger data set
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(Y_test,Y_predict) #On Average Diffrence
```

```python
fig,ax = plt.subplots()
ax.violinplot(Y_predict,showmedians=True,showmeans=True,vert=False); #blue
ax.violinplot(Y_test,showmedians=True,showmeans=True,vert=False); #orange

# ax.yaxis.set_ticks([1, 2, 3])
ax.set_xlabel('Leadership')
ax.set_title('Variation from the estimate')
```
